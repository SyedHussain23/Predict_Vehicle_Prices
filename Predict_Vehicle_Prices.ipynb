## Preprocess the data

### Subtask:
Handle missing values, encode categorical features, and scale numerical features as necessary for model training.

**Reasoning**:
Address missing values, encode categorical features, scale numerical features, and separate the target variable.

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# 1. Address missing values
# Impute numerical features with the median, and categorical features with the most frequent value
numerical_features = ['cylinders', 'mileage', 'doors']
categorical_features = ['description', 'make', 'model', 'engine', 'fuel', 'transmission', 'trim', 'body', 'exterior_color', 'interior_color', 'drivetrain']

# Impute missing 'price' values with the median
df['price'].fillna(df['price'].median(), inplace=True)

numerical_transformer = SimpleImputer(strategy='median')
categorical_transformer = SimpleImputer(strategy='most_frequent')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])

# Apply imputation
df_imputed = preprocessor.fit_transform(df)

# Convert back to DataFrame to handle column names
# Get the names of the imputed numerical columns
imputed_numerical_cols = numerical_features

# Get the names of the imputed categorical columns
# We need to fit the categorical transformer separately to get the feature names
categorical_imputer = SimpleImputer(strategy='most_frequent')
categorical_imputer.fit(df[categorical_features])
imputed_categorical_cols = categorical_features # Imputer does not change feature names for this strategy

# Combine the column names
imputed_cols = imputed_numerical_cols + imputed_categorical_cols

# Create a DataFrame with the imputed data and original column names
df_imputed = pd.DataFrame(df_imputed, columns=imputed_cols, index=df.index)

# Add back the 'year' and 'name' columns which were not imputed in the ColumnTransformer
df_imputed['year'] = df['year']
df_imputed['name'] = df['name']
df_imputed['price'] = df['price'] # Add back the price column

# Reorder columns to match original or desired order
original_columns_order = ['name', 'description', 'make', 'model', 'year', 'price', 'engine', 'cylinders', 'fuel', 'mileage', 'transmission', 'trim', 'body', 'doors', 'exterior_color', 'interior_color', 'drivetrain']
df_imputed = df_imputed[original_columns_order]


# 2. Encode categorical features and 3. Scale numerical features
# Define features again based on the imputed dataframe
numerical_features = ['year', 'mileage', 'cylinders', 'doors'] # 'year' and 'doors' are now included
categorical_features = ['description', 'make', 'model', 'engine', 'fuel', 'transmission', 'trim', 'body', 'exterior_color', 'interior_color', 'drivetrain', 'name']

# Create a column transformer for encoding and scaling
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)],
    remainder='passthrough') # Keep other columns (like 'price')

# Separate features and target
X = df_imputed.drop('price', axis=1)
y = df_imputed['price']

# Apply preprocessing
X_processed = preprocessor.fit_transform(X)

# The output of ColumnTransformer is a sparse matrix or numpy array.
# For simplicity in this step, we'll keep it in this format.
# If needed for later steps, it can be converted back to a DataFrame, but it's not strictly necessary for model training.

display("Processed features (first 5 rows):")
display(X_processed[:5])
display("Target variable (first 5 rows):")
display(y.head())
```

## Feature engineering

### Subtask:
Create new features from existing ones that might be helpful for the model.

**Reasoning**:
Create new features as instructed: extract 'make' from 'name' if not already present (check df_imputed), calculate 'vehicle_age' from 'year', and add a feature for the word count in 'description'. These features will be added to the `df_imputed` DataFrame before the final processing step to be included in `X_processed`. The instructions also mention creating interaction or polynomial features, but based on the previous EDA, a simple linear model might be sufficient as a starting point, and these more complex features can be explored later if necessary. The instructions also mention adjusting the `ColumnTransformer` or adding features after transformation, adding them before the final transformation is the most straightforward approach to ensure they are properly scaled/encoded.

```python
# 1. Extract 'make' from the 'name' column if not already a separate feature.
# Check if 'make' column exists in df_imputed, if not, extract from 'name'
if 'make' not in df_imputed.columns:
    # Assuming the 'name' column starts with the year and then the make
    df_imputed['make'] = df_imputed['name'].apply(lambda x: x.split(' ')[1] if pd.notnull(x) and len(x.split(' ')) > 1 else 'Unknown')

# 2. Create a feature for the age of the vehicle by subtracting the 'year' from the current year (2025).
current_year = 2025
df_imputed['vehicle_age'] = current_year - df_imputed['year']

# 4. Consider creating a feature indicating the number of unique words in the 'description' column
# as a more detailed description might correlate with price.
# Fill NaN descriptions with empty strings before counting words
df_imputed['description'] = df_imputed['description'].fillna('')
df_imputed['description_word_count'] = df_imputed['description'].apply(lambda x: len(set(str(x).split())))

# Redefine features including the new ones for the final processing step
# Assuming 'year' is not needed as 'vehicle_age' is created
numerical_features = ['mileage', 'cylinders', 'doors', 'vehicle_age', 'description_word_count']
# Re-include 'make' in categorical features if it was extracted or already exists
categorical_features = ['description', 'make', 'model', 'engine', 'fuel', 'transmission', 'trim', 'body', 'exterior_color', 'interior_color', 'drivetrain', 'name']


# Create a column transformer for encoding and scaling with updated features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)],
    remainder='passthrough') # Keep other columns (like 'price')

# Separate features and target
X = df_imputed.drop('price', axis=1)
y = df_imputed['price']

# Apply preprocessing with the new features
X_processed = preprocessor.fit_transform(X)

display("Processed features with new features (first 5 rows):")
display(X_processed[:5])
display("Target variable (first 5 rows):")
display(y.head())
```

## Train the model

### Subtask:
Train the selected models (Random Forest Regressor and LightGBM Regressor) on the preprocessed data (`X_processed`) using the target variable (`y`).

**Reasoning**:
Train the selected models (Random Forest Regressor and LightGBM Regressor) on the preprocessed data (`X_processed`) using the target variable (`y`).

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

# Instantiate models
rf_model = RandomForestRegressor(random_state=42)
lgbm_model = lgb.LGBMRegressor(random_state=42)

# Train models
rf_model.fit(X_train, y_train)
lgbm_model.fit(X_train, y_train)

print("Random Forest Model Trained.")
print("LightGBM Model Trained.")
```

## Evaluate the model

### Subtask:
Evaluate the performance of the trained models (Random Forest Regressor and LightGBM Regressor) using appropriate metrics such as Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared.

**Reasoning**:
Evaluate the performance of the trained models using appropriate metrics.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make predictions on the test set
rf_predictions = rf_model.predict(X_test)
lgbm_predictions = lgbm_model.predict(X_test)

# Calculate and print metrics for Random Forest
rf_mae = mean_absolute_error(y_test, rf_predictions)
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)

print("Random Forest Regressor Performance:")
print(f"Mean Absolute Error (MAE): {rf_mae:.2f}")
print(f"Mean Squared Error (MSE): {rf_mse:.2f}")
print(f"R-squared (R2): {rf_r2:.2f}")

# Calculate and print metrics for LightGBM
lgbm_mae = mean_absolute_error(y_test, lgbm_predictions)
lgbm_mse = mean_squared_error(y_test, lgbm_predictions)
lgbm_r2 = r2_score(y_test, lgbm_predictions)

print("\nLightGBM Regressor Performance:")
print(f"Mean Absolute Error (MAE): {lgbm_mae:.2f}")
print(f"Mean Squared Error (MSE): {lgbm_mse:.2f}")
print(f"R-squared (R2): {lgbm_r2:.2f}")

# Briefly compare the performance
print("\nModel Comparison:")
if rf_mae < lgbm_mae:
    print("Random Forest has a lower MAE than LightGBM.")
elif lgbm_mae < rf_mae:
    print("LightGBM has a lower MAE than Random Forest.")
else:
    print("Random Forest and LightGBM have similar MAE.")

if rf_mse < lgbm_mse:
    print("Random Forest has a lower MSE than LightGBM.")
elif lgbm_mse < rf_mse:
    print("LightGBM has a lower MSE than Random Forest.")
else:
    print("Random Forest and LightGBM have similar MSE.")

if rf_r2 > lgbm_r2:
    print("Random Forest has a higher R-squared than LightGBM.")
elif lgbm_r2 > rf_r2:
    print("LightGBM has a higher R-squared than Random Forest.")
else:
    print("Random Forest and LightGBM have similar R-squared.")
```
